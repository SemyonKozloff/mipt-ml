**Вопрос 1**: Опишите, как изменяется адаптивная способность алгоритма ЭСС при изменении параметра $\alpha$ от 0 до 1.

<Ответ>
При $\alpha$ близких 0 адаптивная способность ухудшается: очень низкий вес отдаётся последним отсчётам временного ряда.
При $\alpha$ близких 1 адаптивная способность улучшается: большой вес отдаётся последним отсчётам временного ряда (при $\alpha=1$ прогноз вообще равен последнему известному значению временного ряда).

**Вопрос 2**: Каким следует выбрать параметр сглаживания тренда $\beta$ в модели Хольта (линейный тренд) в случае, когда вы предсказываете временной ряд 1) с плавно меняющимя трендом; 2) стохастически меняющися трендом?

<Ответ>

1. for series with slow changing trend we need to use small beta (sometimes 0), because Simple ES is enough to follow the trend.
2. for stochastic changing trend we need to use biger betas than in precious case because Simple ES is very inert to follow the trend.

**Вопрос 3**: Докажите  эквивалетность формул расчёта (в рекуррентной форме и в форме корректировки прогноза (correction error form) )скрытых компонент временного ряда в модели Тейла-Вейджа (период сезонность обозначается $p$):
$$l_t = \alpha (y_t - s_{t-p}) + (1-\alpha) (l_{t-1} + b_{t-1} ) \color{red}{=} l_{t-1} + b_{t-1} + \alpha e_t;$$
    $$ b_t = \beta (l_{t} - l_{t-1} ) + (1-\beta) b_{t-1} \color{red}{=} b_{t-1} + \alpha\beta e_t;$$
    $$ s_t = \gamma (y_t-l_t) + (1-\gamma) s_{t-p} \color{red}{=} s_{t-p} + \gamma(1-\alpha)e_t.$$

<Ответ>

$$l_t = \alpha (y_t - s_{t-p}) + (1-\alpha) (l_{t-1} + b_{t-1} )=l_{t-1} + b_{t-1} + \alpha e_t;~~~~~~~\text{(3.1)}$$
proof
$$\alpha (y_t - s_{t-p}) -\alpha (l_{t-1} + b_{t-1} )=\alpha e_t;$$
$$y_t - s_{t-p} -l_{t-1} - b_{t-1} =e_t;$$
$$y_t - \hat{y}_{t/t-1} =e_t;$$

$$ b_t = \beta (l_{t} - l_{t-1} ) + (1-\beta) b_{t-1} = b_{t-1} + \alpha\beta e_t;~~~~~~~\text{(3.2)}$$
proof
$$ \beta (l_{t} - l_{t-1} ) -\beta b_{t-1} = \alpha\beta e_t;$$
$$ l_{t} - l_{t-1} - b_{t-1} = \alpha e_t;\quad\text{     see 3.1}$$

$$ s_t = \gamma (y_t-l_t) + (1-\gamma) s_{t-p} = s_{t-p} + \gamma(1-\alpha)e_t.~~~~~~~\text{(3.3)}$$
proof
$$ \gamma (y_t-l_t) -\gamma s_{t-p} = \gamma(1-\alpha)e_t.$$
$$ y_t-l_t -s_{t-p} = (1-\alpha)e_t.$$
$$ y_t-l_t -s_{t-p}- l_{t-1} - b_{t-1} + l_{t-1} + b_{t-1} = (1-\alpha)e_t.$$
$$ (y_t -s_{t-p}- l_{t-1} - b_{t-1})-l_t + l_{t-1} + b_{t-1} = (1-\alpha)e_t.$$
$$ -l_t + l_{t-1} + b_{t-1}  = -\alpha e_t.\quad\text{     see 3.1}$$

**Вопрос 4**: Выведите формулу расчёта скрытых компонент в форме корректировки прогноза (correction error form) временного ряда в модели ЭС с линейным затухающим трендом (WS with additive damped trend):

$$\hat{y}_{t+d|t} = l_t + \left(\phi + \phi^2 + \dots + \phi^{d}\right) b_t.$$,
   
Скрытые компоненты ВР согласно модели в рекуррентной форме записываются:
    $$ l_{t}       = \alpha y_t + \left(1-\alpha\right) \left(l_{t-1} +\phi b_{t-1}\right),$$,
    $$ b_t         = \beta \left(l_t - l_{t-1}\right) + \left(1-\beta\right)\phi b_{t-1}.$$


<Ответ>

$$\hat{y}_{t+d|t} = l_t + \left(\phi + \phi^2 + \dots + \phi^{d}\right) b_t.$$
$$\hat{y}_{t+1|t} = l_t + \phi b_t.$$
$$e_t=y_t-\hat{y}_t=y_t-l_{t-1}-\phi b_{t-1}$$
$$y_t=e_t+l_{t-1}+\phi b_{t-1}$$

Для скрытой компоненты уровня временного ряда $l$:
$$ l_{t}       = \alpha y_t + \left(1-\alpha\right) \left(l_{t-1} +\phi b_{t-1}\right),$$
$$ l_{t}       = \alpha (e_t+l_{t-1}+\phi b_{t-1}) + \left(1-\alpha\right) \left(l_{t-1} +\phi b_{t-1}\right)$$
$$ l_{t}       = \alpha e_t + l_{t-1} +\phi b_{t-1},$$
Для скрытой компоненты тренда  $b$:
$$ b_t         = \beta \left(l_t - l_{t-1}\right) + \left(1-\beta\right)\phi b_{t-1}.$$
$$ b_t         = \beta \left(\alpha e_t +\phi b_{t-1}\right) + \left(1-\beta\right)\phi b_{t-1}.$$
$$ b_t         = \beta \alpha e_t + \phi b_{t-1}.$$

```python
def SimpleExponentialSmoothing(x, h, alpha):
    T = len(x)
    FORECAST = [np.NaN]*(T+h)
    y = x[0]
    for cntr in range(T):
        if not math.isnan(x[cntr]):
            if math.isnan(y):
                y=x[cntr]
            y = y*(1-alpha) + alpha*x[cntr]
            #else do not nothing
        FORECAST[cntr+h] = y
    return FORECAST

def qualityMAPE(x,y):
    qlt = ((x-y).abs()/x).replace([np.inf, -np.inf], np.nan)
    return qlt.mean() , (x-y).abs()


h = 1
start = ts.index[-1]+timedelta(1)
end = ts.index[-1]+timedelta(h)
rng = pd.date_range(start, end)
frc_ts = pd.DataFrame(index = ts.index.append(rng), columns = ts.columns)

# ES params
alpha=np.linspace(0,1,11)[1:-1]

# print(ts.columns)


frc_ts['Real wage'] = SimpleExponentialSmoothing(ts['Real wage'], h, 0.3)


# Example: draw of ts and it's forecast 
print("alpha="+str(0.3)+"\tAccuracy: "+str(1.0-qualityMAPE(ts['Real wage'].loc['1998-09-01':'2015-01-01'],frc_ts['Real wage'].loc['1998-09-01':'2017-01-01'])[0]))
ts['Real wage'].loc['1998-09-01':'2015-01-01'].plot(figsize=(15,5), linewidth=1.0)
frc_ts['Real wage'].loc['1998-09-01':'2017-01-01'].plot(figsize=(15,5), linewidth=1.0)
plt.xlabel("Time ticks")
plt.ylabel("TS values")
plt.legend()
plt.show()

alpha=np.linspace(0,1,21)[1:-1]
score=[]
for alp in alpha:
    
    frc_ts['Real wage'] = SimpleExponentialSmoothing(ts['Real wage'], h, alp)
    score.append(1.0-qualityMAPE(ts['Real wage'].loc['1998-09-01':'2015-01-01'],frc_ts['Real wage'].loc['1998-09-01':'2017-01-01'])[0])
    
plt.plot(alpha,score)
plt.xlabel("alpha")
plt.ylabel("Accuracy")
plt.show()
```

# ANSWER 3.1) and 3.2)

```python
# Start with this code


###################### Winters Exponential Smoothing #########################
# x <array Tx1>- time series, 
# h <scalar> - forecasting delay
# Params <dict> - dictionary with 
#    alpha <scalar in [0,1]> - smoothing parameter
#    delts <scalar in [0,1]> - seasonality smoothing parameter

def WintersExponentialSmoothing(x, h, Params):
    T = len(x)
    alpha = Params['alpha']
    gamma = Params['gamma']
    p = Params['seasonality_period']
    
    FORECAST = [np.NaN]*(T+h)
    
    l= np.NaN
    s= []
    
    for cntr in range(T):
        if not math.isnan(x[cntr]):
            if math.isnan(l):
                l= x[cntr]
            if len(s)==0:
                for i in range(p):
                    s.append(x[i])
            if cntr<p:
                l = alpha*(x[cntr]-s[cntr])+(1-alpha)*l # recurrent smoothing of level 
            else:
                s.append(gamma*(x[cntr]-l)+(1-gamma)*s[cntr-p])
                l = alpha*(x[cntr]-s[cntr-p])+(1-alpha)*l # recurrent smoothing of level 
            
        FORECAST[cntr+h] = l + s[cntr-p+1]
    return FORECAST
```

```python
h = 1
rng = pd.date_range(ts.index[-1]+timedelta(1), ts.index[-1]+timedelta(h))

w_frc_ts = pd.DataFrame(index = ts.index.append(rng), columns = ts.columns)

alpha=np.linspace(0,1,21)[1:-1]
score=[]
for alp in alpha:
    
    w_frc_ts['Real wage'] = WintersExponentialSmoothing(ts['Real wage'], h, {'alpha':alp, 'gamma':0.5, 'seasonality_period':12})
    score.append(1.0-qualityMAPE(ts['Real wage'].loc['1998-09-01':'2015-01-01'],w_frc_ts['Real wage'].loc['1998-09-01':'2017-01-01'])[0])
    
plt.plot(alpha,score)
plt.xlabel("alpha")
plt.ylabel("Accuracy")
plt.show()

gamma=np.linspace(0,1,21)[1:-1]
score=[]
for gam in gamma:
    
    w_frc_ts['Real wage'] = WintersExponentialSmoothing(ts['Real wage'], h, {'alpha':0.6, 'gamma':gam, 'seasonality_period':12})
    score.append(1.0-qualityMAPE(ts['Real wage'].loc['1998-09-01':'2015-01-01'],w_frc_ts['Real wage'].loc['1998-09-01':'2017-01-01'])[0])
    
plt.plot(gamma,score)
plt.xlabel("gamma")
plt.ylabel("Accuracy")
plt.show()
```

The best alpha and gamma for 98-15yy are (0.62, 0.42)

```python
# Start with this code


###################### Winters Exponential Smoothing #########################
# x <array Tx1>- time series, 
# h <scalar> - forecasting delay
# Params <dict> - dictionary with 
#    alpha <scalar in [0,1]> - smoothing parameter
#    delts <scalar in [0,1]> - seasonality smoothing parameter

def TWExponentialSmoothing(x, h, Params):
    T = len(x)
    alpha = Params['alpha']
    beta = Params['beta']
    gamma = Params['gamma']
    p = Params['seasonality_period']
    
    FORECAST = [np.NaN]*(T+h)
    
    l= np.NaN
    b=np.NaN
    s= []
    
    for cntr in range(T):
        if not math.isnan(x[cntr]):
            if math.isnan(l):
                l= x[cntr]
            if math.isnan(b):
                b= 0
            
            if len(s)==0:
                for i in range(p):
                    s.append(x[i])
                    
                    
            if cntr<p:
                l_old=l
                l = alpha*(x[cntr]-s[cntr])+(1-alpha)*(l+b)
                b=beta*(l-l_old)+(1-beta)*b
            else:
                l_old=l
                s.append(gamma*(x[cntr]-l)+(1-gamma)*s[cntr-p])
                l = alpha*(x[cntr]-s[cntr-p])+(1-alpha)*(l+b) # recurrent smoothing of level 
                b=beta*(l-l_old)+(1-beta)*b
            
        FORECAST[cntr+h] = l+b + s[cntr-p+1]
    return FORECAST
```

```python
h = 1
rng = pd.date_range(ts.index[-1]+timedelta(1), ts.index[-1]+timedelta(h))

tw_frc_ts = pd.DataFrame(index = ts.index.append(rng), columns = ts.columns)

alpha=np.linspace(0,1,21)[1:-1]
score=[]
for alp in alpha:
    
    tw_frc_ts['Real wage'] = TWExponentialSmoothing(ts['Real wage'], h, {'alpha':alp, 'gamma':0.5, 'beta':0, 'seasonality_period':12})
    score.append(1.0-qualityMAPE(ts['Real wage'].loc['1998-09-01':'2015-01-01'],tw_frc_ts['Real wage'].loc['1998-09-01':'2017-01-01'])[0])
    
plt.plot(alpha,score)
plt.xlabel("alpha")
plt.ylabel("Accuracy")
plt.show()

gamma=np.linspace(0,1,21)[1:-1]
score=[]
for gam in gamma:
    
    tw_frc_ts['Real wage'] = TWExponentialSmoothing(ts['Real wage'], h, {'alpha':0.6, 'gamma':gam, 'beta':0, 'seasonality_period':12})
    score.append(1.0-qualityMAPE(ts['Real wage'].loc['1998-09-01':'2015-01-01'],tw_frc_ts['Real wage'].loc['1998-09-01':'2017-01-01'])[0])
    
plt.plot(gamma,score)
plt.xlabel("gamma")
plt.ylabel("Accuracy")
plt.show()

beta=np.linspace(0,1,21)[1:-1]
score=[]
for bet in beta:
    
    tw_frc_ts['Real wage'] = TWExponentialSmoothing(ts['Real wage'], h, {'alpha':0.6, 'gamma':0.5, 'beta':bet, 'seasonality_period':12})
    score.append(1.0-qualityMAPE(ts['Real wage'].loc['1998-09-01':'2015-01-01'],tw_frc_ts['Real wage'].loc['1998-09-01':'2017-01-01'])[0])
    
plt.plot(beta,score)
plt.xlabel("beta")
plt.ylabel("Accuracy")
plt.show()
```

```python
def MulExponentialSmoothing(x, h, Params):
    T = len(x)
    alpha = Params['alpha']
    beta = Params['beta']
    gamma = Params['gamma']
    p = Params['seasonality_period']
    
    FORECAST = [np.NaN]*(T+h)
    
    l= np.NaN
    b=np.NaN
    s= []
    
    for cntr in range(T):
        if not math.isnan(x[cntr]):
            if math.isnan(l):
                l= x[cntr]
            if math.isnan(b):
                b= 0
            
            if len(s)==0:
                s.append(1)
                for i in range(1,p):
                    s.append(x[i]/x[i-1])
                    
                    
            if cntr<p:
                l_old=l
                l = alpha*(x[cntr]/s[cntr])+(1-alpha)*(l+b)
                b=beta*(l-l_old)+(1-beta)*b
            else:
                l_old=l
                s.append(gamma*(x[cntr]/l)+(1-gamma)*s[cntr-p])
                l = alpha*(x[cntr]-s[cntr-p])+(1-alpha)*(l+b) # recurrent smoothing of level 
                b=beta*(l-l_old)+(1-beta)*b
            
        FORECAST[cntr+h] = (l+b)*s[cntr-p+1]
    return FORECAST
```

```python
h = 1
rng = pd.date_range(ts.index[-1]+timedelta(1), ts.index[-1]+timedelta(h))

mw_frc_ts = pd.DataFrame(index = ts.index.append(rng), columns = ts.columns)

alpha=np.linspace(0,1,21)[1:-1]
score=[]
for alp in alpha:
    
    mw_frc_ts['Real wage'] = MulExponentialSmoothing(ts['Real wage'], h, {'alpha':alp, 'gamma':0.5, 'beta':0, 'seasonality_period':12})
    score.append(1.0-qualityMAPE(ts['Real wage'].loc['1998-09-01':'2015-01-01'],mw_frc_ts['Real wage'].loc['1998-09-01':'2017-01-01'])[0])
    
plt.plot(alpha,score)
plt.xlabel("alpha")
plt.ylabel("Accuracy")
plt.show()

gamma=np.linspace(0,1,21)[1:-1]
score=[]
for gam in gamma:
    
    mw_frc_ts['Real wage'] = MulExponentialSmoothing(ts['Real wage'], h, {'alpha':0.99, 'gamma':gam, 'beta':0, 'seasonality_period':12})
    score.append(1.0-qualityMAPE(ts['Real wage'].loc['1998-09-01':'2015-01-01'],mw_frc_ts['Real wage'].loc['1998-09-01':'2017-01-01'])[0])
    
plt.plot(gamma,score)
plt.xlabel("gamma")
plt.ylabel("Accuracy")
plt.show()

beta=np.linspace(0,1,21)[1:-1]
score=[]
for bet in beta:
    
    mw_frc_ts['Real wage'] = MulExponentialSmoothing(ts['Real wage'], h, {'alpha':0.6, 'gamma':0.5, 'beta':bet, 'seasonality_period':12})
    score.append(1.0-qualityMAPE(ts['Real wage'].loc['1998-09-01':'2015-01-01'],mw_frc_ts['Real wage'].loc['1998-09-01':'2015-01-01'])[0])
    
plt.plot(beta,score)
plt.xlabel("beta")
plt.ylabel("Accuracy")
plt.show()
```
