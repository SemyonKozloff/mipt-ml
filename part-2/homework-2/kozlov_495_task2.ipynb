**Вопрос 4**: Покажите, что для модели  
    $$AR(p)\colon \;\;\; y_t = \phi_1 y_{t-1} + \phi_2 y_{t-2} + \dots + \phi_p y_{t-p} + \varepsilon_t,$$**  
    где $y_t$ стационарный процесс с нулевым средним,$~~$
    $\phi_1,\dots,\phi_p$ - константы ($\phi_p \neq 0$), $~~$
    $\varepsilon_t$ несмещённый гауссовский шум с дисперсией $\sigma_\varepsilon^2 $
выолнено:

   * a) частный коэффициент автокорреляции
   $$\phi_{\tau\tau} = \mathsf{E}[y_t-\hat y_t, y_{t+\tau} - \hat y_{t+\tau}]= \begin{cases}0 &, \tau>p; \\ f(\tau,p)>0 &, 0\leq\tau\leq p. \end{cases}$$
	
  где $\hat y_t$ - прогноз $y_t$ согласно следующей регрессионной модели $$y_t = \sum_{j=1}^{\tau-1}\beta_j y_{t+j}$$
    $\hat y_{t+\tau}$ - forecast of $y_{t\tau}$ with following regression model $$y_{t+\tau} = \sum_{j=1}^{\tau-1}\gamma_j y_{t+j}.$$

$$ y_{t+\tau} - \hat y_{t+\tau}=\varepsilon_{t+\tau} + \sum_{i=1}^{p}\phi_iy_{t+\tau-i} -  \sum_{j=1}^{\tau-1}\gamma_j y_{t+j} = \varepsilon_{t+\tau} + \sum_{i=1}^{p}\phi_iy_{t+\tau-i} -  \sum_{j=1}^{\tau-1}\gamma_j y_{t+j}$$


$$S = \sum_{i=1}^{p}\phi_iy_{t+\tau-i} -  \sum_{j=1}^{\tau-1}\gamma_j y_{t+j} = \sum_{i=1}^{p}\phi_iy_{t+\tau-i} -  \sum_{i=1}^{\tau-1}\gamma_{\tau-i} y_{t+\tau-i}$$

$$\phi_i = \gamma_{\tau-i} \;\; i = 1..p$$

$$\gamma_{\tau-i}=0 \;\;\; i=(p+1)..(\tau-1)$$

if $p \leq\tau-1$ $(p <\tau)$ then $S = 0$ 
then.
    $$\phi_{\tau\tau} = \mathsf{E}[(y_t - \hat y_t)( \varepsilon_t + S)] = 0 $$

if $p \geq \tau$ then
$$\phi_{\tau\tau} \ne 0 $$
$$\mathsf{E}[y_i y_j] \ne 0 \;\;\; |i-j| \leq p $$

```python
ts6['val'] = ts6['item 65672']
ts = ts6.loc[:, ['val']]
ts.columns = ['val']
ts = ts.resample('D').mean()
ts = ts.fillna(ts['val'].mean())
ts.plot(figsize=(15,5))
plt.show()
```

Посмотрим, что можно сделать со стационарностью ряда.

```python
k = int(len(ts) * 0.8)
train_ts = ts.iloc[:k, :]
test_ts = ts.iloc[k:, :]
```

```python
lam = 0.1

ts['val_boxcox'] = (np.array(ts['val']) ** lam + 1) ** (1 / lam) 
ts['val_boxcox'].plot(figsize=(15,5))
plt.show()

kpss_stat = stattools.kpss(np.array(ts['val_boxcox']))
print('diff 1 The KPSS test statistic %f, \npvalue: %.2f \nmaximal Lag: %i \ntable distribution: %s' % 
      (kpss_stat[0], kpss_stat[1], kpss_stat[2], kpss_stat[3]))
df_stat = stattools.adfuller(np.array(ts['val_boxcox']))
print('The Augmented Dickey-Fuller test statistic %f, \npvalue: %.2f \nmaximal Lag: %i \ncritical values: %s' % 
      (df_stat[0], df_stat[1], df_stat[2], df_stat[4]))
```

Видим, что по критерию Дики-Фуллер ряд является стационарным. Будем брать d = 0.

```python
res = sm.tsa.seasonal_decompose(ts['val'].fillna(0))
resplot = res.plot()
```

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(train_ts.values.squeeze(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(train_ts.values.squeeze(), lags=40, ax=ax2)
```

Попробуем q равным 1, 2 или 4, p равным 1 или 2.

```python
def arima_fit(*pdq):
    return sm.tsa.ARIMA(train_ts, (*pdq)).fit()

def arima_fit_aic(*pdq):
    try:
        aic = arima_fit(pdq).aic
        print('ARIMA', pdq, '->', aic)
        return aic
    except:
        return 1e100
```

```python
from itertools import product
pdq_cands = product([1, 2, 4], [0], [1, 2])

aics = [(arima_fit_aic(*par), par) for par in pdq_cands]
```

```python
best_pdq = min(aics)[1]
min(aics)
```

Мда уж.

```python
model = arima_fit(best_pdq)
model.resid.plot(figsize=(10,4))
plt.show()

df_stat = stattools.adfuller(np.array(model.resid))
print('The Augmented Dickey-Fuller test statistic %f, \npvalue: %.2f \nmaximal Lag: %i \ncritical values: %s' % 
      (df_stat[0], df_stat[1], df_stat[2], df_stat[4]))

stat = stats.wilcoxon(model.resid.values)
print('WILCOXON Statistic: %f\npvalue: %.2f' % (stat[0], stat[1]))
```

```python
test_ts['val_forecast'], test_ts['val_err'] = model.forecast(steps=len(test_ts))[:2]
ts['val_forecast'] = test_ts['val_forecast']
ts['val_forecast_err'] = test_ts['val_err']

def plot_forecast(test):
    plt.figure(figsize=(14, 6))
    plt.plot(ts.index, ts['val'])
    plt.plot(ts.index, ts['val_forecast'], color='red')

    def plus_minus(a, b):
        return (a - b, a + b)

    plt.fill_between(ts.index, *plus_minus(ts['val_forecast'], ts['val_forecast_err']), alpha=0.3, color='red')
    plt.show()
    
    print('MACAPE: %.2f' % qualityMACAPE(ts['val_forecast'], ts['val'])[0])
    
plot_forecast(ts)
```

Получается полная фигня.
